```python
df = pd.read_csv('/home/vagrant/notebooks/fall_2014_lessons/datasets/challenger_oring.csv',
                 parse_dates=0, names=['date', 'temp', 'failure'], header=False).dropna()
df
```

```python
df.failure = df.failure.astype(int)
```

```python
df.temp = df.temp.astype(int)
```

```python
X = df.temp.values
y = df.failure.values
```

```python
X
```

```python
y
```

```python
fig, axes = plt.subplots(1,1, figsize=(12,8))
axes.set_ylim((-0.2, 1.2))
plt.plot(X, y, 'x')
```

```python
model = LinearRegression()
model.fit(X.reshape(-1,1), y)
X_test = np.linspace(31., 20., 90., 10).reshape(-1,1)
y_hat = model.predict(X_test)
```

```python
fig, axes = plt.subplots(1,1, figsize=(12, 8))
axes.set_ylim((-0.2, 1.2))
axes.set_xlim((50,85))
plt.plot(X, y, 'o')
plt.plot(X_test, y_hat)
```
