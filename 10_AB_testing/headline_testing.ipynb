```python
def plot_beta(views, ctr):
    upper = beta.ppf(1.0-0.0001, views*ctr, views*(1-ctr))
    xs = linspace(0,upper,2000)
    ys = [beta.pdf(x, views*ctr, views*(1-ctr)) for x in xs]
    ys /= (sum(ys) / ((1.0*len(ys))))
    plot(xs,ys)
```

```python

def demonstrate(a_views, b_views, a_ctr, b_ctr, size):
    pb = percent_better(a_views, b_views, a_ctr, b_ctr, size)
    
    if a_ctr > b_ctr:
        winner = 'A'
    else:
        winner = 'B'
        pb = 1.0 - pb
    
    plot_beta(a_views, a_ctr)
    hold(True)
    plot_beta(b_views, b_ctr)
    legend(['a','b']);
    title('Headline %s is better in %.2f%% of the cases' % (winner, pb*100.0));
    xlabel('CTR')
    xlim(0,0.25)


    
def percent_better(a_views, b_views, a_ctr, b_ctr, size):
    ra = beta.rvs(a_views*a_ctr, a_views*(1-a_ctr), size=(size))
    rb = beta.rvs(b_views*b_ctr, b_views*(1-b_ctr), size=(size))
    return sum(ra >= rb) / (1.0*size)
```

```python
fig = figure(figsize=(10,5))
demonstrate(100,200, 0.04969, 0.13287, size=1000000)
```
